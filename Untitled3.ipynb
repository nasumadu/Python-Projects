# Make the Position into Four (4) Categories
Goalkeeper (GK)
Defender (DF)
Midfielder (MF)
Forward (FD)

```python
#Forming multiple positions into four positional groups

def position_conversion(value):
    
    if value == 'RF' or value == 'ST' or value == 'LF' or value == 'RS' or value == 'LS' or value == 'CF' or value == 'LW' or value == 'RW':
        return 'F'
        
    elif   value == 'RCM' or value == 'LCM' or value == 'LDM' or value == 'CAM' or value == 'CDM' or value == 'RM' \
         or value == 'LAM' or value == 'LM' or value == 'RDM' or value == 'CM' or value == 'RAM':
        return 'M'
    
    elif value == 'RCB' or value == 'CB' or value== 'LCB' or value == 'LB' or value == 'RB' or value == 'RWB' or value == 'LWB':
        return 'D'
    
    else:
        return value
fifa1['Position'] = fifa1['Position'].apply(position_conversion)
```

```python
#Position Number count 

fifa1['Position'] = fifa1['Position'].apply(position_conversion)
fifa1['Position'].value_counts()
```

# Predicting Top Young GK's Overall

```python
# Splitting the dataset into a Training & Testing Set
split = np.random.rand(len(fifa_gk)) < 0.85
train = fifa_gk[split]
test = fifa_gk[~split]
```

```python
#Creating a Linear Regression Model
regr = linear_model.LinearRegression()
```

```python
# Training & fitting the model
fifa_gk_x_train = train[['Age', 'Potential' ]]
fifa_gk_y_train = train[['Overall']]
regr.fit(fifa_gk_x_train, fifa_gk_y_train)
```

```python
# Testing the model
fifa_gk_x_test = test[['Age', 'Potential']]
fifa_gk_y_test = test[['Overall']]
```

```python
fifa_gk_pred =regr.predict(fifa_gk_x_test)
```

```python
r2_score(fifa_gk_y_test, fifa_gk_pred)
```

```python
sqrt(mean_squared_error(fifa_gk_y_test, fifa_gk_pred))
```

```python
fifa_gk_x_test['Predicted Ratings'] = fifa_gk_pred
```

```python
fifa_gk = pd.merge(fifa_gk,fifa_gk_x_test)
fifa_gk.head(20)
```
